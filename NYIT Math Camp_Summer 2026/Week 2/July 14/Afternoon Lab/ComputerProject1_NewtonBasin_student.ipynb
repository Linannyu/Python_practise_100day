## Background

This morning you learned that Newton's method finds a root of $f(x)=0$ by iterating

$$p_n = p_{n-1} - \frac{f(p_{n-1})}{f'(p_{n-1})}, \quad n \geq 1$$

starting from an initial guess $p_0$.

A natural question: **does the choice of $p_0$ matter?**

For a function with *multiple roots*, different starting points can lead to *different roots* — sometimes in surprising ways. The set of all starting points that lead to a given root is called the **basin of attraction** of that root.

In this project you will:
1. Implement Newton's method in Python.
2. Apply it to 41 different starting points.
3. Color each starting point by which root it converges to.
4. Visualize the result — your first Newton basin diagram!

## Step 2 — Implement Newton's method

Recall Algorithm 1.2 from the lecture:

```
INPUT p0, TOL, NI
STEP 1: set i = 1
STEP 2: while i <= NI:
    STEP 3: p = p0 - f(p0) / f'(p0)
    STEP 4: if |p - p0| < TOL → STOP (converged)
    STEP 5: i = i + 1
    STEP 6: p0 = p
STEP 7: output failure message
```

We wrap this in a Python function that **returns two things**:
- `p`         — the final approximation
- `converged` — `True` if the method converged, `False` if it ran out of iterations or failed

We also add one safety check: if $|f'(p_0)|$ is extremely small (near zero), the next
step $p_0 - f(p_0)/f'(p_0)$ would blow up — so we stop and return `False`.

## Step 6 — Investigate: what went wrong at $x_0 = 1.4$ and $x_0 = 2.6$?

The two black diamonds are near the *critical points* of $f$ — where $f'(x) \approx 0$.

Recall that Newton's step is $p = p_0 - \dfrac{f(p_0)}{f'(p_0)}$.

**What happens when $f'(p_0)$ is very small?**

````python
# Compute f'(x) at the two problematic starting points
for x0 in [1.4, 2.6]:
    print(f"x0 = {x0:.1f}  |  f'(x0) = {df(x0):.6f}  |  f(x0)/f'(x0) = {f(x0)/df(x0):.4f}")

print()
# Find where f'(x) = 0 exactly
# f'(x) = 3x^2 - 12x + 11 = 0  =>  x = (12 ± sqrt(144 - 132)) / 6
disc = 144 - 4*3*11
c1 = (12 - disc**0.5) / 6
c2 = (12 + disc**0.5) / 6
print(f"f'(x) = 0 at x = {c1:.4f} and x = {c2:.4f}")
print(f"Our problem points x0=1.4 and x0=2.6 are very close to these!")
````

When $f'(p_0)$ is small, the Newton step $f(p_0)/f'(p_0)$ is **large** —
the method jumps far away and then needs many more iterations to settle down.
With only 10 iterations maximum, it runs out of time.

**✏️ Experiment:** What happens if you increase `NI` to 20? Do the black diamonds disappear?

````python
# Try NI = 20 for the problematic points
for x0 in [1.4, 2.6]:
    p10, c10 = newton(f, df, x0, NI=10)
    p20, c20 = newton(f, df, x0, NI=20)
    print(f"x0={x0:.1f}  |  NI=10: converged={c10}, p={p10:.8f}  "
          f"|  NI=20: converged={c20}, p={p20:.8f}")
````

## Step 7 — The surprising crossing: why does $x_0 = 1.5$ converge to root 3?

Geometrically, $x_0 = 1.5$ is closest to root $x_1 = 1$ and $x_2 = 2$, yet it converges to $x_3 = 3$.
Let's trace exactly what happens, iteration by iteration.

````python
# Trace Newton's method step by step for x0 = 1.5
print(f"{'Iter':>4}  {'p':>18}  {'f(p)':>14}  {'f\'(p)':>12}  {'step size':>12}")
print("-" * 68)

p0 = 1.5
for i in range(1, 11):
    fp0  = f(p0)
    dfp0 = df(p0)
    if abs(dfp0) < 1e-15:
        print("  Derivative too small — stopping.")
        break
    p    = p0 - fp0 / dfp0
    step = abs(p - p0)
    print(f"{i:>4}  {p:>18.12f}  {f(p):>14.6e}  {dfp0:>12.6f}  {step:>12.6e}")
    if step < 1e-12:
        print(f"  → Converged to {p:.10f}")
        break
    p0 = p
````

> **Observe:** At iteration 1, Newton makes a very large jump.
> This is because $f'(1.5)$ is small (we are near the local maximum of $f$),
> so the step size $f(p_0)/f'(p_0)$ is large — launching the method all the way
> past root 2 and toward root 3.

This is the **key insight** of the Newton basin: the geometry of $f'(x)$ — not just the
position of the starting point — determines which root you reach.
